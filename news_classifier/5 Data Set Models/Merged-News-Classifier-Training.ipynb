```python
import pandas as pd
df = pd.read_csv("../data/5TrainingData.csv")
df
```

```python
df.drop(columns=["_id"], inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
label = le.fit_transform(df["category"])

df.drop(columns=["category"], inplace=True)

df["category"] = label

df
```

```python
import en_core_web_lg

nlp = en_core_web_lg.load()

def nlpify(text):
    text = str(text)
    """tokenized_text = nlp(text)
    token_list = []
    for token in tokenized_text:
        token_list.append(token.text)
    
    final_text = []
    for word in token_list:
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False:
            final_text.append(word)
    fin = " ".join(final_text)
    #a = nlp(text)
    #b = a.vector"""
    return nlp(text).vector

df["headline"] = df["headline"].apply(nlpify)

#word = "word"
#print(nlp(word).vector)
#print(word.vector)

#df4.head()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X = df["headline"]
y = df["category"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline


logisticregression = LogisticRegression()
logisticregression.fit(X_train_2d, y_train)
```

```python
print("Logistic Regression Classifier Accuracy: " + str(round((logisticregression.score(X_test_2d, y_test)*100), 2)) + "%")
```

```python
from sklearn.metrics import classification_report

y_pred = logisticregression.predict(X_test_2d)

classification_repo = classification_report(y_true=y_test, y_pred=y_pred, target_names=["BUSINESS", "ENTERTAINMENT",  "HEALTH", "SCIENCE", "SPORTS", "TECH", "U.S. NEWS", "WORLD NEWS"])
print(classification_repo)
```

```python
logisticregression = LogisticRegression(class_weight="balanced")
logisticregression.fit(X_train_2d, y_train)

y_pred = logisticregression.predict(X_test_2d)

classification_repo = classification_report(y_true=y_test, y_pred=y_pred, target_names=["BUSINESS", "ENTERTAINMENT", "HEALTH", "SCIENCE", "SPORTS", "TECH", "U.S. NEWS", "WORLD NEWS"])
print(classification_repo)
```
