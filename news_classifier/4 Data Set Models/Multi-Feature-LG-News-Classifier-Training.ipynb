```python
import pandas as pd
df = pd.read_csv("../data/4-TrainingData.csv")
df
```

```python
df.drop(columns=["_id", "link", "short_description", "authors", "date"], inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
label = le.fit_transform(df["category"])

df.drop(columns=["category"], inplace=True)

df["category"] = label

df
```

```python
import en_core_web_lg

nlp = en_core_web_lg.load()

def lemmify(text):
    text = str(text)
    tokenized_text = nlp(text)
    lemmatized_text = ' '.join([token.lemma_ for token in tokenized_text])
    return nlp(lemmatized_text).vector
    
df["lemmatizedheadline"] = df["headline"].apply(lemmify)
```

```python
def nlpify(text):
    text = str(text)
    """tokenized_text = nlp(text)
    token_list = []
    for token in tokenized_text:
        token_list.append(token.text)
    
    final_text = []
    for word in token_list:
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False:
            final_text.append(word)
    fin = " ".join(final_text)
    #a = nlp(text)
    #b = a.vector"""
    return nlp(text).vector

df["headline"] = df["headline"].apply(nlpify)

#word = "word"
#print(nlp(word).vector)
#print(word.vector)

#df4.head()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X = pd.concat([df.drop(["category"], axis=1)])
X = [list1 + list2 for list1, list2 
                   in zip(df["headline"], 
                          df["lemmatizedheadline"])]
y = df["category"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

linearsvc = LinearSVC()
linearsvc.fit(X_train_2d, y_train)

logisticregression = LogisticRegression()
logisticregression.fit(X_train_2d, y_train)

multinomialnb = Pipeline([('scaler', MinMaxScaler()),('clf', MultinomialNB())])
multinomialnb.fit(X_train_2d, y_train)
```

```python
print("LinearSVC Classifier Accuracy: " + str(round((linearsvc.score(X_test_2d, y_test)*100), 2)) + "%")
print("Logistic Regression Classifier Accuracy: " + str(round((logisticregression.score(X_test_2d, y_test)*100), 2)) + "%")
print("MultinomialNB Classifier Accuracy: " + str(round((multinomialnb.score(X_test_2d, y_test)*100), 2)) + "%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

kneighborsclassifier = KNeighborsClassifier()
kneighborsclassifier.fit(X_train_2d, y_train)

randomforestclassifier = RandomForestClassifier()
randomforestclassifier.fit(X_train_2d, y_train)

print("KNeighbors Classifier Accuracy: " + str(round((kneighborsclassifier.score(X_test_2d, y_test)*100), 2)) + "%")
print("RandomForest Classifier Accuracy: " + str(round((randomforestclassifier.score(X_test_2d, y_test)*100), 2)) + "%")
```

```python
from sklearn.metrics import classification_report

y_pred = logisticregression.predict(X_test_2d)

classification_repo = classification_report(y_true=y_test, y_pred=y_pred, target_names=["BUSINESS", "ENTERTAINMENT",  "HEALTH", "SCIENCE", "SPORTS", "TECH", "U.S. NEWS", "WORLD NEWS"])
print(classification_repo)
```

```python
logisticregression = LogisticRegression(class_weight="balanced")
logisticregression.fit(X_train_2d, y_train)

y_pred = logisticregression.predict(X_test_2d)

classification_repo = classification_report(y_true=y_test, y_pred=y_pred, target_names=["BUSINESS", "ENTERTAINMENT", "HEALTH", "SCIENCE", "SPORTS", "TECH", "U.S. NEWS", "WORLD NEWS"])
print(classification_repo)
```
