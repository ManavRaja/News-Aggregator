```python
##### Categories to classify into: BUSINESS, ENTERTAINMENT, HEALTH, SCIENCE, SPORTS, TECH, U.S. NEWS, WORLD NEWS

###### Contains data from "RishabhMisraHuffPostDataset", "GoogleNewsDataset", "AGNewsDataset" without Science/Tech, "MicrosoftNewsDatset" only health category

###### BUSINESS - 45344
###### Entertainment - 24644
###### HEALTH - 25875
###### POLITICS - 35602
###### SCIENCE - 6565
###### SPORTS - 45207
###### TECH - 9038
###### Total - 200278
```

```python
import pandas as pd
df = pd.read_pickle("../data/9TrainingData.pkl")
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
label = le.fit_transform(df["category"])

df.drop(columns=["category"], inplace=True)

df["category"] = label
df
```

```python
from sklearn.model_selection import train_test_split

X = df["headline"]
y = df["category"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logisticregression = LogisticRegression()
logisticregression.fit(X_train_2d, y_train)

y_pred = logisticregression.predict(X_test_2d)

classification_repo = classification_report(y_true=y_test, y_pred=y_pred, target_names=["BUSINESS", "ENTERTAINMENT", "HEALTH", "POLITICS", "SCIENCE", "SPORTS", "TECH"])
print(classification_repo)
```

```python
logisticregression2 = LogisticRegression(class_weight="balanced")
logisticregression2.fit(X_train_2d, y_train)

y_pred2 = logisticregression2.predict(X_test_2d)

classification_repo2 = classification_report(y_true=y_test, y_pred=y_pred2, target_names=["BUSINESS", "ENTERTAINMENT", "HEALTH", "POLITICS", "SCIENCE", "SPORTS", "TECH"])
print(classification_repo2)
```
